```python
import pandas as pd
import sklearn as sk
import numpy as np
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print(cancer.keys())
print(cancer['feature_names'])
```

```python
cancer_features = cancer['data']
cancer_label = cancer['target']

features = [feature.replace(' ', '_') for feature in cancer['feature_names']]
print(features)
cancerPandas = pd.DataFrame(data= np.c_[cancer_features, cancer_label],
                     columns= features + ['target'])
cancerDF = spark.createDataFrame(cancerPandas)
display(cancerDF)

```
